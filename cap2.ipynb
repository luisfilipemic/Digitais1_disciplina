## 2.1 Sistema Decimal

O sistema decimal é composto por um conjunto de 10 algarismos, sendo de 0 a 9;

cada posição possui um peso/magnitude equivalente.

por exemplo:

23

2: Dezenas
3: Unidades

2*10+3*1 = 23;

- Exemplo:
Expresse o número decimal 47 como uma soma dos valores de cada dígito.

teremos:
$$ (4\cdot 10^1) + (7\cdot 10^0) = 47$$

Determine o valor de cada dígito em 939.

3° digito = $$9\cdot 10^2 = 900$$
2° digito = $$ 3\cdot 10^1 = 30$$
1° digito = $$9\cdot 10^0 = 9$$

Expresse o número decimal 568,23 como uma soma dos valores de cada dígito

teremos:

$$ 568,23 = (5*10^2) + (6*10^1) + (8*10^0) + (2*10^{-1}) + (3*10^{-2}) = 568,23$$

## Números binários 

Números binários é um sistema de numeração com peso 2, diferente dos decimais.
ou seja, cada digito possui seu respectivo peso $2^n$.

| NÚMERO DECIMAL | NÚMERO BINÁRIO |
|----------------|----------------|
| 0              | 0000           |
| 1              | 0001           |
| 2              | 0010           |
| 3              | 0011           |
| 4              | 0100           |
| 5              | 0101           |
| 6              | 0110           |
| 7              | 0111           |
| 8              | 1000           |
| 9              | 1001           |
| 10             | 1010           |
| 11             | 1011           |
| 12             | 1100           |
| 13             | 1101           |
| 14             | 1110           |
| 15             | 1111           |

Essa tabela é básica para compreensão de binario para decimal, perceba que é de 4 BITS, porém podemos contar $n$ números.

- Exemplo:

com cinco BITS podemos contar até $2^n -1$

Fórmula geral:

$$Qtds de numeros para bits = 2^n-1$$

Por que -1?

Por que existe o zero sendo contado.

#### Estrutura de pesos.

Em um binario cada bit possui respectivo peso.
O bit mais esquerda é considerado o menos sisgnificativo (LSB), e a direita o mais significativo(MSB);

Numeros binarios com virgula:

$$2^{n–1}... 2^{3} 2^{2} 2^{1} 2^{0}, 2^{–1} 2^{–2}... 2^{–n}$$

Onde a virgula é denotada como virgula binária;

- Exemplo 2-3:

teremos: 

$$(2*10^0) + (2*10^2) + (2*10^3) + (2*10^5) + (2*10^6) = 109 $$


- Exemplo 2-4:

## Números siinalizados:

- Bit de sinal:

0 - positivo 

1 - negativo

#### FOrma sinal magnitude


Na forma sinal-magnitude, um número negativo tem os mesmos bits de magnitude co
mo o número positivo correspondente mas o bit de sinal é 1 em vez de zero.

Exemplo:

+25 = -25 

a magnitude será igual, porém alteramos somente o bit de sinal de 0 para 1;



#### Complemento de 1:

COmplemento de 1 para numeros positivos:

- igualmente o sinal magnitude positivo;


COmplemento de 1 para numeros negativos:

- complementow de 1 para o sinal magnitude.

ex:

-25 = comp1(+25)

+25 = sinalmag(+25)

#### Complemento de 2:

COmplemento de 2 para numeros positivos:

- igualmente o sinal magnitude positivo;


COmplemento de 2 para numeros negativos:

- complementow de 2 para o sinal magnitude.

ex:

-25 = comp2(+25)

+25 = sinalmag(+25)

#### O valor decimal de numeros sinalizados(processo inverso):

- SInal magnitude:

no sinal magnitude basta somar os pesos relativos aos bits 1s;

o bit de sinal informará se é positivo ou negativo;

- Complemento de 1:

Positivos:

Para os positivos basta somar os respectivos pesos 1s

Negastivos:
Os
valores decimais de números negativos são determinados atribuindo um valor negativo ao peso do
bit de sinal, somando os pesos relativos aos bits 1s e somando 1 ao resultado

- COmplemento de dois:

POsitivos:

Basta somar os respectivos pesos dos bits 1s

negativos:

basta somar o peso dos bits 1s, sendo que o bit sinalizado terá peso negativo na soma;


Observa-se que o complemento de 2 é mais fácil;


#### Faixa de representação:

Bom, a faixa de representação de numeros sinalizados varia conforme a quantidade de bits;

formula geral para o numero de combinação:


$$Faixa = −(2^{n−1})a +(2^{n−1} − 1)$$
