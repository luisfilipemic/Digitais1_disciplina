## Formas de onda digitais:

A forma de onda digital presente no sistema é o PULSO, que corresponde a dois niveis logicos, representados por 0 e 1.


















```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definindo os dados (tempo e níveis)
# Queremos um pulso que sobe em t=2 e desce em t=5
tempo =  [0, 1, 2, 3, 4, 5, 6, 7]
niveis = [0, 0, 1, 1, 1, 0, 0, 0]

# 2. Criando o gráfico
plt.figure(figsize=(10, 4))

# O parâmetro 'where="post"' garante que o nível mude exatamente no ponto definido
plt.step(tempo, niveis, where='post', color='blue', linewidth=2)

# 3. Estilização para parecer um osciloscópio/diagrama lógico
plt.title('Pulso Digital (Níveis 0 e 1)')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (Lógica)')
plt.ylim(-0.2, 1.2)  # Dá um respiro acima e abaixo para ver melhor
plt.yticks([0, 1])    # Mostra apenas os níveis 0 e 1 no eixo Y
plt.grid(True, linestyle='--', alpha=0.7)

# Exibir
plt.show()
```

conforme o gráfico podemos ver carteristicas como a **borda de subida e borda de descida**.

observação.:
o pulso descrito pelo grafico é ideal, no sistema real o tempo de comutação ou inversão de nivel logico ocorre num instante infinitesimal, logo, não é instantaneo.

## Operações lógicas binárias:

ou portas logicas

quando pensamos em operações logicas binarias devemos pensar nas principais portas logicas.

são elas:

1. not 
2. and
3. or
4. nad
5. nor

**NOT**:

a operação not tem como carteristica uma mudança logica na saída.

possui uma entrada e um saída.

condições:

| entrada | saída |


|-----------------|


| alto    | baixo |


| baixo   | alto  |

chamamos de porta logica barrada.








**AND**:

possui duas entradas e uma saída.

a saída será verdadeira se e  somente se todas as entradas forem verdadeiras.


**OR** 
possui duas entradas e uma saída.

A saída terá nivel logico alto se qualquer entrada possuir nivel logico alto.

exemplo.: 
se há uma entrada alto e outra baixo, a saída será alto.

## Circuitos integrados de funções fixas (CI).

circuitos integrados possuem intenamente um conjunto de funções logicas ou portas logicas.

construção:
- chip
- ecapsulamento de plastico
- pinos/ terminais


**tipos de CIs**

1. PTH = ci montado por pinos na placa (sigla pin)

2. SMD = ci montada na superfice da placa(sufarce)

vantagem:
economia de espaço e versatibilidade
menores

**Numeração dos pinos**

verificar pela parte fronta do ci o indicador do pino 1.

**Classificação de complexidade do cis(determinados com a quantidade de circuitod integrados)**:

- SSI(small scale integration):
 até dez circuitos de porta
- MSI(medium scale integration):
10 a 100 portas
- LSI(large scale integration):
100 a 10.000
- VLSI(very large scale integration):
10.000 a 100.000 portas
