### Operação e expressões boleanas.

## Adição boleana:

A adição boleana pode ser implementada usando portas logicas or:

portanto teremos que a saída de uma or pode ser difinida como:

$$ X = A + B$$

## Multiplicação boleana:

A multiplicação boleana pode ser implementada usando portas and.

Portanto, teremos:

$$ X = A\cdot B$$

POrtanto, a saída será 1 se e somente se, todas as entradas forem numericamente iguais a 1.

### Leis e regras da álgebra boleana:

## Propiedades fundamentais: 

- Comutativa:

$$A + B = B + A$$

$$AB = BA$$

- Associativa:

$$A + (B +C) = (A + B) +C$$

$$(AB)C = A(BC)$$

- Distributiva:

A lei distributiva tambem é válida para todos os casos.


## Regras da algebra boleana:



Prova real: P10:

$$A + AB = A(B + 1) = A(1) = A$$

### Análise boleana de circuitos lógicos:

Expressão Booleana para um Circuito Lógico:
Para obter a expressão Booleana para um dado circuito lógico, comece pelas entradas mais à es
querda e, percorrendo o circuito até a saída final, escreva a expressão para cada porta lógica.

## Construindo uma Tabela-verdade para um Circuito Lógico:

a Tabela da expressão boleana pode ser feita a partir dos valores das variaveis que dão resposta alta no circuito logico.

### Formas padronizadas de expressões boleanas:

## Forma padronizada soma de produtos:

Em soma de produtos teremos a seguinte estrutura:

$$AB + ABC + BCD$$

Em que $ABCD$ é o dominio da expressão.

- Como deixar a forma soma de produtos padronizadas:

note que na expressão abaixo, o dominio não é representado de forma completa em cada termo produto:

$$AB + ABC + BCD$$

Neste caso é necessário multiplicar cada termo produto por 1.

nesse caso usaremos o teorema :

$$A + Abarrado = 1$$

## Representação Binária de um Termo-produto Padrão

Seja a expressão:

$$ABCD$$

Terá resposta alta qual todo o dominio for igual a 1.
